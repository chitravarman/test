```python
scraper=sntwitter.TwitterSearchScraper("#elon musk")
tweets=[]
for i, tweet in enumerate(scraper.get_items()):
    data=[
     tweet.date,
     tweet.id,
     tweet.content,
     tweet.user.username,
     tweet.likeCount,
     tweet.retweetCount,
     tweet.url,
     ]
    tweets.append(data)
    if i>1000:
        break
tweets

    
```

```python
df=pd.DataFrame(
    tweets, columns=["date", "id", "content", "username", "like_count", "retweet", "Url",]
)
df
```
